```python
%matplotlib inline

import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
path = '112914.csv'

df = pd.read_csv(path, header=0, low_memory=False, names=["id","tweet","tid","uid","datetime", "coordinates"], index_col='datetime', parse_dates=True)
```

```python
df.head()
```

```python
print min(df.index.date)
print max(df.index.date)
```

```python
new_df = df.drop(['id','tid','uid'], 1)
```

```python
wanted_range = pd.date_range(start='2014-11-08',end='2014-11-28')
wanted_range
```

```python
new_df = new_df.groupby(df.index.date).count().reindex(wanted_range)
new_df = new_df.drop(['coordinates'], 1)
```

```python
new_df
```

```python
new_df['tweet'].sum()
```

```python
new_df.plot(kind="line", rot=25, figsize=(20, 12), alpha=.7, color='Green', legend=False, linewidth=3)

plt.title('Tweet Volume: "Palestine" -- November 2014')
plt.ylabel('# of tweets')
```

PLOTTING VELOCITY

```python
vel = []

index = 0
for t in new_df.tweet[1:]:
    vel.append(float(t) / float(new_df.tweet[index]))
    index+=1
```

```python
vel
```

```python
velocity_range = pd.date_range(start='2014-11-09',end='2014-11-28')

vel_df = pd.DataFrame({ 'vel': vel, 'date': velocity_range})
```

```python
vel_df = vel_df.set_index(['date'])
```

```python
vel_df.plot(kind="line", rot=25, figsize=(15, 10), alpha=.7, color='DarkGreen', legend=False)
```
